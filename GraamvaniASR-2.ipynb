```python
import re

def remove_ids_from_file(file_path):
    # Read the contents of the file
    with open(file_path, 'r', encoding='utf-8') as file:
        transcriptions = file.readlines()

    # Remove IDs from each transcription
    cleaned_transcriptions = []
    for transcription in transcriptions:
        cleaned_transcription = re.sub(r'^\d{2}-\d{5}-\d{2}\s+', '', transcription)
        cleaned_transcriptions.append(cleaned_transcription.strip())

    # Write the cleaned transcriptions back to the file
    with open('reference.txt', 'w', encoding='utf-8') as file:
        for cleaned_transcription in cleaned_transcriptions:
            file.write(cleaned_transcription + '\n')

remove_ids_from_file("./gramvaani/testsets/gramvaani/text")
```

```python
def calculate_wer_whisper(original_transcription_file, whisper_transcription_file):
  """
  This function calculates WER using Whisper's command-line tool (if installed).

  Args:
      original_transcription_file (str): Path to the file containing reference transcripts.
      whisper_transcription_file (str): Path to the file containing Whisper's predictions.

  Returns:
      float: The WER (Word Error Rate) or None if an error occurs.
  """

  command = [ "wer", original_transcription_file, whisper_transcription_file]

  try:

    process = subprocess.run(command, capture_output=True, check=True)
    output = process.stdout.decode()


    lines = output.strip().splitlines()[1:]


    if not lines:
        print("Error: WER output seems empty or missing WER value.")
        return None


    wer_value_str = lines[0].strip().split()[1]
    wer_value = float(wer_value_str[:-1])

    return wer_value

  except subprocess.CalledProcessError as e:
    print(f"Error: WER calculation failed ({e})")
    return None

```

```python

original_transcripts_path = "/content/original_transcription.txt"
whisper_transcripts_path = "/content/whisper_transcription.txt"

wer_score = calculate_wer_whisper(original_transcripts_path, whisper_transcripts_path)

if wer_score is not None:
  print(f"Word Error Rate (WER): {wer_score:.4f}")
else:
  print("Error: WER calculation failed.")
```
